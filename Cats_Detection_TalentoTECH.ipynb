# IMPORT LIBS

```python
import cv2
from google.colab.patches import cv2_imshow

import numpy as np
```

# EXERCICE CAT FACE

```python
faceClassif = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
```

```python
# Load image
image = cv2.imread('six_cats.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Cat faces detection
faces = faceClassif.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    maxSize=(200, 200)
)
print(faces)
```

```python
# Paint ractangles around of detected faces
for (x, y, w, h) in faces:
  cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
# Show images with highlights edges
print('Objetos detectados')
cv2_imshow(image)
```

# MEJORAR DETECCION CON PARAMETROS

```python
# Cat faces detection
faces_2 = faceClassif.detectMultiScale(
    gray,
    scaleFactor=1.01,
    minNeighbors=2,
    minSize=(80, 80),
    maxSize=(150, 150)
)
print(faces_2)
```

```python
# Reload image
image = cv2.imread('six_cats.jpg')

# Paint ractangles around of detected faces
for (x, y, w, h) in faces_2:
  cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
# Show images with highlights edges
print('Objetos detectados')
cv2_imshow(image)
```
