```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataset
# Let´s build the values lists
name_list = ['The Legend of Valor', 'Cybernetic Siege', 'Galactic Conquest', 'Fantasy Quest IX', 'Super Sports Showdown']

year_list = [2019, 2022, 2018, 2023, 2021]
plataform_list = ['PC', 'PlayStation', 'Xbox', 'Nintendo Switch', 'PC']
dev_list = ['Blaze Studios', 'FutureTech Games', 'Nova Interactive', 'Mythic Games', 'GamePro Sports']
genre_list = ['Adventure', 'Shooter', 'Strategy', 'Rol', 'Sports']
subgenre_list = ['Action', 'Cyber-punk', 'RTS', 'Action', 'Soccer']
copies_list = [500000, 8000000, 2000000, 3000000, 10000000]
value_list = [200, 250, 170, 300, 120] # Dollars
concurrent_list = [12000, 350000, 80000, 90000, 500000]


# Let´s use a dictionary ({key:value})
# with the format {"str":list}
data_dict = {"name":name_list,
             "year":year_list,
             "plataform":plataform_list,
             "developers":dev_list,
             "genre":genre_list,
             "subgenre":subgenre_list,
             "copies":copies_list,
             "value":value_list,
             "concurrent":concurrent_list,
             }

#print(data_dict)

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

#print(df)
# Call a variable
df
```
